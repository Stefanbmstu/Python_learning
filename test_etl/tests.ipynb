 Реимпорт - обновление колонок из фронта (current_columns) на актуальные колонки из базы данных (actual_columns), но при совпадении имен перезаписываются все параметры колонок, кроме key (для сохранения связей (relations) в рабочем отображении между старыми колонками), в остальных случаях создаем ключ сами. Если в actual_columns уже нет какой-то колонки, которая есть в current_columns, то она, соотвественно, удаляется.

```python
import uuid

# Определение тестовых фикстур для current_table и actual_columns
current_table = {
		"id_table": "a0fb4a9a-28b8-4e19-8d0c-90a2c21e3081",
		"name": "classes",
		"columns": [
			{
				"name": "class",
				"type": "varchar",
				"is_primary_key": 0,
				"length": -1,
				"key": "a8b52312-3406-47c6-b5c6-99129a9f8bf6",
				"props": {
					"input": "true",
					"output": "true"
				}
			},
			{
				"name": "type",
				"type": "varchar",
				"is_primary_key": 0,
				"length": -1,
				"key": "e29f7469-f019-4cd3-a7e3-bc89b4b036b3",
				"props": {
					"input": "true",
					"output": "true"
				}
			},
			{
				"name": "country",
				"type": "varchar",
				"is_primary_key": 0,
				"length": -1,
				"key": "bedf2de9-6eb4-4a87-a686-8a84578eb68f",
				"props": {
					"input": "true",
					"output": "true"
				}
			},
			{
				"name": "numguns",
				"type": "integer",
				"is_primary_key": 0,
				"length": -1,
				"key": "e6bc51c4-4f7c-41cc-aa05-4fc44d8a04f9",
				"props": {
					"input": "true",
					"output": "true"
				}
			},
			{
				"name": "bore",
				"type": "double_precision",
				"is_primary_key": 0,
				"length": -1,
				"key": "63875565-e41a-47bf-9166-f79942161d0f",
				"props": {
					"input": "true",
					"output": "true"
				}
			},
			{
				"name": "displacement",
				"type": "integer",
				"is_primary_key": 0,
				"length": -1,
				"key": "7974c20a-7874-4e73-80d9-3806565b56d8",
				"props": {
					"input": "true",
					"output": "true"
				}
			}
		],
		"type": "Postgres",
		"meta": {
			"real_name": "classes"
		},
		"position": {
			"x": "null",
			"y": "null"
		},
		"size": {
			"width": 200,
			"height": 273
		},
		"connector_id": 2,
		"type_table": "source"
	}

actual_columns = [
        {
            "name_column": "class",
            "type_column": "varchar",
            "is_primary_key": False,
            "length": -1,
            "precision": -1,
            "scale": -1,
            "props": {
                "input": "true",
                "output": "true"
            }
        },
        {
            "name_column": "type",
            "type_column": "varchar",
            "is_primary_key": False,
            "length": -1,
            "precision": -1,
            "scale": -1,
            "props": {
                "input": "true",
                "output": "true"
            }
        },
        {
            "name_column": "country",
            "type_column": "varchar",
            "is_primary_key": False,
            "length": -1,
            "precision": -1,
            "scale": -1,
            "props": {
                "input": "true",
                "output": "true"
            }
        },
        {
            "name_column": "numguns",
            "type_column": "integer",
            "is_primary_key": False,
            "length": -1,
            "precision": -1,
            "scale": -1,
            "props": {
                "input": "true",
                "output": "true"
            }
        },
        
        {
            "name_column": "displacement",
            "type_column": "integer",
            "is_primary_key": False,
            "length": -1,
            "precision": -1,
            "scale": -1,
            "props": {
                "input": "true",
                "output": "true"
            }
        }
    ]

current_columns = current_table["columns"]

current_columns_dict = {col["name"]: col for col in current_columns}

# Список имён актуальных колонок для проверки существования
actual_column_names = [col["name_column"] for col in actual_columns]

for col in actual_columns:
    # Перебираем колонки из актуальных данных 'actual_columns'
    col_name = col["name_column"]
    if col_name in current_columns_dict:
        # Обновляем параметры существующей колонки, кроме 'key'
        for key, value in col.items():
            if key != "key":
                # Убираем суффикс '_column' и обновляем значение
                current_columns_dict[col_name][key.replace(
                    "_column", "")] = value
    else:
        # Добавляем новую колонку с уникальным 'key'
        new_col = {key.replace(
            "_column", ""): value for key, value in col.items() if key != "key"}
        new_col["key"] = str(uuid.uuid4())
        current_columns.append(new_col)

 # Фильтруем current_columns, чтобы удалить колонки, которых нет в actual_columns
filtered_current_columns = [col for col in current_columns if col["name"] in actual_column_names]

updated_table = filtered_current_columns

print(updated_table)
```
